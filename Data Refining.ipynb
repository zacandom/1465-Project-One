```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Load merged dataset
df = pd.read_csv("CES_Data_merged_wave9_16.csv")

# Outcome variable: cons_tot only exists in wave 16
df['logSpend'] = np.log(df['cons_tot_w16']) * 100

# Treatments: fill missing with 0
treat_cols = [f'treat{i}_w9' for i in range(1,6)]
df[treat_cols] = df[treat_cols].fillna(0)

# Priors: fill missing with 0
priormean_col = 'egrea_imean_bt_w9'
prioruncert_col = 'egrea_istd_bt_w9'
df[priormean_col] = df[priormean_col].fillna(0)
df[prioruncert_col] = df[prioruncert_col].fillna(0)

# Interactions: treatment * prior
for i in range(1,6):
    df[f'treat{i}_x_mean'] = df[f'treat{i}_w9'] * df[priormean_col]
    df[f'treat{i}_x_uncert'] = df[f'treat{i}_w9'] * df[prioruncert_col]

# Controls: fill missing columns with 0 if not present
control_cols = ['dedu2_w9','dedu3_w9','age_w9','hhsize_w9','lhhnetinc_w9',
                'liquid_w9','male_w9','k1010_3_w9','pr2010_w9','pr2110_w9',
                'cnt1_w9','cnt2_w9','cnt3_w9','cnt4_w9','cnt5_w9','cnt6_w9']

for col in control_cols:
    if col not in df.columns:
        df[col] = 0
df[control_cols] = df[control_cols].fillna(0)

# Regression matrix
X = df[treat_cols + 
       [f'treat{i}_x_mean' for i in range(1,6)] + 
       [f'treat{i}_x_uncert' for i in range(1,6)] + 
       control_cols]

# Add intercept
X = sm.add_constant(X)

# Fit OLS with robust SE
model = sm.OLS(df['logSpend'], X)
results = model.fit(cov_type='HC3')

# Create summary table for LaTeX
summary_table = pd.DataFrame({
    'Variable': results.params.index,
    'Coefficient': results.params.values,
    'Std. Error': results.bse,
    't stat': results.tvalues,
    'p value': results.pvalues
})

# Export to LaTeX
latex_table = summary_table.to_latex(index=False, float_format="%.3f",
                                     caption="OLS Regression Results",
                                     label="tab:ols_results")

with open("OLS_summary_table.tex", "w") as f:
    f.write(latex_table)

print("LaTeX table saved as OLS_summary_table.tex")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import ElasticNet

# Load your dataset
df = pd.read_csv("CES_Data_merged_wave9_16.csv")

# Outcome
y = np.log(df['cons_tot_w16'].fillna(0)) * 100

# Predictors (fill missing values with 0)
predictors = ['treat1_w9','treat2_w9','treat3_w9','treat4_w9','treat5_w9',
              'egrea_imean_bt_w9','egrea_istd_bt_w9',
              'dedu2_w9','dedu3_w9','age_w9','hhsize_w9','lhhnetinc_w9',
              'liquid_w9','male_w9','k1010_3_w9',
              'cnt1_w9','cnt2_w9','cnt3_w9','cnt4_w9','cnt5_w9','cnt6_w9']

X = df[predictors].fillna(0).values

# Standardize predictors
scaler = StandardScaler()
Xs = scaler.fit_transform(X)

# Define lambda grid
lambdas = 10**np.linspace(8, -2, 100) / y.std()

# Fit Ridge path (ElasticNet with l1_ratio=0)
coef_list = []
for alpha in lambdas:
    ridge = ElasticNet(alpha=alpha, l1_ratio=0, fit_intercept=True, max_iter=10000)
    ridge.fit(Xs, y)
    coef_list.append(ridge.coef_)

# Convert to array
coef_array = np.array(coef_list)

# Plot coefficient paths
plt.figure(figsize=(10, 7))
for i in range(coef_array.shape[1]):
    plt.plot(-np.log10(lambdas), coef_array[:, i], label=predictors[i])
plt.xlabel(r'$-\log_{10}(\lambda)$', fontsize=14)
plt.ylabel('Standardized Coefficients', fontsize=14)
plt.title('Ridge Coefficient Paths', fontsize=16)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler

# --- 1. Prepare data ---
# Keep only rows with positive cons_tot_w16
df_clean = df[df['cons_tot_w16'] > 0].copy()

# Outcome variable (log of consumption * 100)
y = np.log(df_clean['cons_tot_w16']) * 100

# Predictor variables
predictors = [
    'treat1_w9','treat2_w9','treat3_w9','treat4_w9','treat5_w9',
    'egrea_imean_bt_w9','egrea_istd_bt_w9',
    'dedu2_w9','dedu3_w9','age_w9','hhsize_w9','lhhnetinc_w9',
    'liquid_w9','male_w9','k1010_3_w9',
    'cnt1_w9','cnt2_w9','cnt3_w9','cnt4_w9','cnt5_w9','cnt6_w9'
]

X = df_clean[predictors].fillna(0).values

# Standardize predictors
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 2. Fit Ridge regressions across a grid of lambdas ---
lambdas = np.logspace(-3, 3, 100)  # from 0.001 to 1000
coefs = []

for a in lambdas:
    ridge = Ridge(alpha=a, fit_intercept=True)
    ridge.fit(X_scaled, y)
    coefs.append(ridge.coef_)

coefs = np.array(coefs)

# --- 3. Plot Ridge coefficient paths ---
plt.figure(figsize=(10, 6))
for i, col in enumerate(predictors):
    plt.plot(lambdas, coefs[:, i], label=col)

plt.xscale('log')
plt.xlabel("Lambda (α)")
plt.ylabel("Coefficient value")
plt.title("Ridge Coefficient Paths")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler

# --- 1. Prepare data (same as before) ---
df_clean = df[df['cons_tot_w16'] > 0].copy()
y = np.log(df_clean['cons_tot_w16']) * 100

predictors = [
    'treat1_w9','treat2_w9','treat3_w9','treat4_w9','treat5_w9',
    'egrea_imean_bt_w9','egrea_istd_bt_w9',
    'dedu2_w9','dedu3_w9','age_w9','hhsize_w9','lhhnetinc_w9',
    'liquid_w9','male_w9','k1010_3_w9',
    'cnt1_w9','cnt2_w9','cnt3_w9','cnt4_w9','cnt5_w9','cnt6_w9'
]

X = df_clean[predictors].fillna(0).values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 2. Define groups ---
variables_of_interest = [
    'treat1_w9','treat2_w9','treat3_w9','treat4_w9','treat5_w9',
    'egrea_imean_bt_w9','egrea_istd_bt_w9'
]
controls = [v for v in predictors if v not in variables_of_interest]

# --- 3. Fit Ridge regressions across lambdas ---
lambdas = np.logspace(-3, 3, 100)
coefs = []
for a in lambdas:
    ridge = Ridge(alpha=a, fit_intercept=True)
    ridge.fit(X_scaled, y)
    coefs.append(ridge.coef_)
coefs = np.array(coefs)

# --- 4. Plot Ridge coefficient paths ---
plt.figure(figsize=(10, 6))

# Plot controls in black
for i, col in enumerate(predictors):
    if col in controls:
        plt.plot(lambdas, coefs[:, i], color="black", linewidth=1, alpha=0.7)

# Plot variables of interest with distinct colors
colors = plt.cm.tab10(np.linspace(0, 1, len(variables_of_interest)))
for i, col in enumerate(predictors):
    if col in variables_of_interest:
        plt.plot(lambdas, coefs[:, i], label=col, color=colors[variables_of_interest.index(col)], linewidth=2)

plt.xscale("log")
plt.xlabel("Lambda (α)")
plt.ylabel("Coefficient value")
plt.title("Ridge Coefficient Paths")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import KFold

# --- 1. Define lambdas (alphas) ---
lambdas = np.logspace(-3, 3, 100)

# --- 2. Cross-validation setup ---
cv = KFold(n_splits=10, shuffle=True, random_state=42)

# --- 3. RidgeCV automatically does CV over the alphas ---
ridge_cv = RidgeCV(alphas=lambdas, store_cv_values=True, cv=cv)
ridge_cv.fit(X_scaled, y)

# cv_values_ has shape (n_samples, n_alphas)
mse_mean = np.mean(ridge_cv.cv_values_, axis=0)
mse_std = np.std(ridge_cv.cv_values_, axis=0) / np.sqrt(len(y))

# --- 4. Plot ---
plt.figure(figsize=(8, 6))
plt.semilogx(lambdas, mse_mean, marker="o", label="Mean CV MSE")
plt.fill_between(lambdas, mse_mean - mse_std, mse_mean + mse_std, alpha=0.2)

plt.axvline(ridge_cv.alpha_, color="red", linestyle="--", label=f"Best λ = {ridge_cv.alpha_:.3f}")

plt.xlabel("Lambda (α)")
plt.ylabel("Mean Squared Error (CV)")
plt.title("Ridge Cross-Validation Curve")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.model_selection import KFold, cross_val_score

# --- 1. Define lambdas (alphas) ---
lambdas = np.logspace(-3, 3, 100)

# --- 2. Cross-validation setup ---
cv = KFold(n_splits=10, shuffle=True, random_state=42)

# --- 3. Compute CV MSE for each lambda ---
mse_mean, mse_std = [], []
for alpha in lambdas:
    ridge = Ridge(alpha=alpha)
    scores = cross_val_score(ridge, X_scaled, y, cv=cv, scoring="neg_mean_squared_error")
    mse = -scores  # flip sign because sklearn returns negative MSE
    mse_mean.append(mse.mean())
    mse_std.append(mse.std() / np.sqrt(len(scores)))

mse_mean = np.array(mse_mean)
mse_std = np.array(mse_std)

# --- 4. Find best lambda ---
best_idx = np.argmin(mse_mean)
best_alpha = lambdas[best_idx]

# --- 5. Plot ---
plt.figure(figsize=(8, 6))
plt.semilogx(lambdas, mse_mean, marker="o", label="Mean CV MSE")
plt.fill_between(lambdas, mse_mean - mse_std, mse_mean + mse_std, alpha=0.2)

plt.axvline(best_alpha, color="red", linestyle="--", label=f"Best λ = {best_alpha:.3f}")

plt.xlabel("Lambda (α)")
plt.ylabel("Mean Squared Error (CV)")
plt.title("Ridge Cross-Validation Curve")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import lasso_path
from sklearn.preprocessing import StandardScaler

# --- 1. Prepare data ---
df_clean = df[df['cons_tot_w16'] > 0].copy()
y = np.log(df_clean['cons_tot_w16']) * 100

predictors = [
    'treat1_w9','treat2_w9','treat3_w9','treat4_w9','treat5_w9',
    'egrea_imean_bt_w9','egrea_istd_bt_w9',
    'dedu2_w9','dedu3_w9','age_w9','hhsize_w9','lhhnetinc_w9',
    'liquid_w9','male_w9','k1010_3_w9',
    'cnt1_w9','cnt2_w9','cnt3_w9','cnt4_w9','cnt5_w9','cnt6_w9'
]
X = df_clean[predictors].fillna(0).values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 2. Compute LASSO path ---
alphas_lasso, coefs_lasso, _ = lasso_path(X_scaled, y, alphas=np.logspace(-3, 3, 100))

# --- 3. Set up variable groups ---
vars_interest = ['treat1_w9','treat2_w9','treat3_w9','treat4_w9','treat5_w9',
                 'egrea_imean_bt_w9','egrea_istd_bt_w9']
colors_interest = plt.cm.tab10(np.linspace(0, 1, len(vars_interest)))  # distinct colors
color_map = {}

for i, var in enumerate(predictors):
    if var in vars_interest:
        color_map[var] = colors_interest[vars_interest.index(var)]
    else:
        color_map[var] = 'black'

# --- 4. Plot coefficient paths ---
plt.figure(figsize=(10, 6))
for i, var in enumerate(predictors):
    plt.plot(-np.log(alphas_lasso), coefs_lasso[i].T, 
             label=var if var in vars_interest else None,
             color=color_map[var], linewidth=1.5 if var in vars_interest else 1,
             alpha=0.9 if var in vars_interest else 0.6)

plt.xlabel("-log(Lambda)")
plt.ylabel("Coefficient Value")
plt.title("LASSO Coefficient Paths")

# Show only variables of interest in legend
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, labels, loc="best")

plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LassoCV
from sklearn.model_selection import KFold

# --- 1. Prepare data (same as before) ---
X = df_clean[predictors].fillna(0).values
y = np.log(df_clean['cons_tot_w16']) * 100
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 2. Fit LassoCV ---
cv = KFold(n_splits=5, shuffle=True, random_state=42)
lasso_cv = LassoCV(alphas=np.logspace(-3, 3, 100), cv=cv, max_iter=10000)
lasso_cv.fit(X_scaled, y)

# --- 3. Extract MSE for each alpha ---
mse_path = np.mean(lasso_cv.mse_path_, axis=1)  # average over folds
std_path = np.std(lasso_cv.mse_path_, axis=1) / np.sqrt(lasso_cv.mse_path_.shape[1])  # SE

# --- 4. Plot ---
plt.figure(figsize=(10, 6))
plt.errorbar(-np.log(lasso_cv.alphas_), mse_path, yerr=std_path, fmt='-o', capsize=3)
plt.xlabel("-log(Lambda)")
plt.ylabel("CV Mean Squared Error")
plt.title("LASSO Cross-Validation Curve")
plt.tight_layout()
plt.show()
```
